```python
# function to remove columns with 'FA'

def remove_columns_with_FA(df):
    columns_to_keep = [col for col in df.columns if 'FA' not in col]
    df_cleaned = df[columns_to_keep]
    return df_cleaned


```

```python
col_copy = nutri['alim_nom_eng'].copy()
nutri['KEY_WORDS'] = col_copy
nutri.head()
```

```python
def move_column(df, col_name, new_position):
    cols = df.columns.tolist()
    cols.remove(col_name)
    cols.insert(new_position, col_name)
    return df[cols]

# changement d'index dans le DF pour la colonne "KEY_WORDS" pour un confort de lecture
nutri = move_column(nutri, 'KEY_WORDS', 5)
nutri.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Fonction de recherche
def search_glucides(keyword, df):
    # Encoder le mot-clé en embedding
    keyword_embedding = model.encode([keyword])
    print(f"Embedding du mot-clé '{keyword}': {keyword_embedding}")

    # Calculer la similarité cosinus entre l'embedding du mot-clé et les embeddings dans la base de données
    similarities = cosine_similarity(keyword_embedding, np.array(nutri['key_words_embedding'].tolist()))
    print(f"Similarités : {similarities}")

    # Trouver l'index du mot-clé le plus similaire
    best_match_index = np.argmax(similarities)
    print(f"Index du mot-clé le plus similaire : {best_match_index}")

    # Récupérer les valeurs de glucides associées
    result = nutri.iloc[best_match_index]['Carbohydrate (g/100g)']
    print(f"Valeur de glucides associée : {result}")

    return result

# Mots-clés à tester
keywords = ["pineapple"]

# Tester la fonction
for keyword in keywords:
    print(f"\nTest pour le mot-clé '{keyword}':")
    glucides = search_glucides(keyword, nutri)
    print(f"Les glucides pour '{keyword}' sont : {glucides}")
```
