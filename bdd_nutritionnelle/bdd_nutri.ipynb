```python
import pandas as pd
import re



col_copy = nutri['alim_nom_eng'].copy()

# Supprimer les mots terminant par "ed" (en split par virgule ou espace)
def remove_ed_words(texte):
    mots = re.split(r'[,\s]+', texte)  # split sur virgule ou espace
    mots_filtres = [mot for mot in mots if not re.search(r'ed$', mot)]
    return ','.join(mots_filtres)

#Appliquer la fonction à la copie
nutri['KEY_WORDS'] = col_copy.apply(remove_ed_words)

nutri.head()
```

```python
import en_core_web_sm
nlp = en_core_web_sm.load()

# Import Stop words
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
nutri['KEY_WORDS'] = nutri['KEY_WORDS'].str.replace('/', ' ', regex=False)
nutri.head()
```

```python
nutri['KEY_WORDS'] = nutri['KEY_WORDS'].str.replace(',', ' ', regex=False)
nutri.head()
```

```python
nutri['KEY_WORDS'] = nutri['KEY_WORDS'].str.replace('-', ' ', regex=False)

nutri['KEY_WORDS'] = nutri['KEY_WORDS'].fillna('').apply(lambda x: x.lower())
nutri.head()
```

```python
nutri['KEY_WORDS'].apply(
    lambda text: ' '.join([word for word in text.split() if word.lower() not in STOP_WORDS])
)
nutri.head()
```
